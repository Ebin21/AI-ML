```python
sales["Total"] = pd.to_numeric(sales["Total"], errors='coerce')
```

```python
cols_to_convert = ['Total', 'Quantity', 'Price']
for col in cols_to_convert:
    sales[col] = pd.to_numeric(sales[col], errors='coerce')
```

```python
sales.mean(numeric_only=True)["Total"]
```

```python
location_list = sales.groupby('Location')
location=[x for x,y in location_list]
location
```

```python
sales.groupby('Location').sum(numeric_only=True)['Total']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(location, sales.groupby('Location').sum(numeric_only=True)['Total'])
```

```python
plt.pie(sales.groupby('Location').sum(numeric_only=True)['Total'], labels=location, autopct='%1.1f%%')
```

```python
location_sales = sales.groupby(['Gender', 'Location']).count()['Invoice ID']
location_sales
```

```python
unstacked_location_sales = location_sales.unstack(level=0) 
```

```python
unstacked_location_sales.plot(kind='bar')
```

```python
sales['Day'] = pd.to_datetime(sales['Date']).dt.day
sales.head(10)
```

```python
day_sales = sales.groupby('Day').sum(numeric_only=True)['Total']
day_sales.plot(kind='line')
```
